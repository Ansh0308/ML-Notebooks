```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy.io import loadmat
mat =loadmat("/content/ex7data2.mat")
X=mat["X"]
```

```python
#initialize the centroids
import random
init_cent=random.sample(range(0,len(X)),3)
```

```python
init_cent
```

```python
centroids=[]
for i in init_cent:
  centroids.append(X[i])
centroids
```

```python
def calc_dis(x1, x2):
    x1 = np.array(x1)
    x2 = np.array(x2)
    return np.sqrt(np.sum((x1 - x2)**2))
```

```python
def findclosestCentroid(centroid,X):
  assi_cent=[]
  for i in X:
    distance=[]
    for j in centroid:
      distance.append(calc_dis(i,j))
    assi_cent.append(np.argmin(distance))
  return assi_cent

```

```python
findclosestCentroid(centroids,X)
```

```python
#mean of the observation falling in each cluster category
def new_centroid(centroid,X):
  new_centroids=[]
  new_df=pd.concat([pd.DataFrame(X),pd.DataFrame(centroid,columns=['centroid'])],axis=1)
  for c in set(new_df['centroid']):
    current_cluster=new_df[new_df['centroid']==c][new_df.columns[:-1]]
    new_mean=current_cluster.mean(axis=0)
    new_centroids.append(new_mean)
  return new_centroids
```

```python
for i in range(10):
  obs_cen=findclosestCentroid(centroids,X)
  centroids=new_centroid=(obs_cen,X)
  plt.figure()
  plt.scatter(np.array(centroids)[:,0],np.array(centroids)[:,1],color='black')
  plt.scatter(X[:,0],X[:,1])
  plt.show()
```
