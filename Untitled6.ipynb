```python
import smtplib
from email.mime.text import MIMEText
import random

def send_otp(receiver_email):
    # Gmail SMTP setup
    smtp_server = "smtp.gmail.com"
    smtp_port = 587
    sender_email = "ansh.raythatha122354@marwadiuniversity.ac.in"
    app_password = "rhuu zjry pvxf qhvg"  # generated from Google

    # Generate a random 6-digit OTP
    otp = str(random.randint(100000, 999999))

    # Create email content
    msg = MIMEText(f"Your OTP for Engineer's Day registration is: {otp}")
    msg["Subject"] = "OTP Verification - ICT Engineer's Day"
    msg["From"] = sender_email
    msg["To"] = receiver_email

    # Send email
    with smtplib.SMTP(smtp_server, smtp_port) as server:
        server.starttls()
        server.login(sender_email, app_password)
        server.sendmail(sender_email, [receiver_email], msg.as_string())

    print(f"✅ OTP sent to {receiver_email}")
    return otp

send_otp("ansh.raythatha122354@marwadiuniversity.ac.in")
```
