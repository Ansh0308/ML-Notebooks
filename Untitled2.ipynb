```python
hobby_counts=data['Hobby'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(hobby_counts, labels=hobby_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Do people find coding as a hobby?')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

print(f"Counts for Hobby: \n{hobby_counts}")
```

```python
open_source_contributors = data[data['OpenSource'] == 'Yes'].shape[0]
total_respondents_open_source_question = data['OpenSource'].count() # count non-NaN responses

print(f"Number of people contributing to open-source projects: {open_source_contributors}")
print(f"Total respondents for Open Source question: {total_respondents_open_source_question}")
print(f"Percentage of respondents contributing to open source: "
      f"{(open_source_contributors / total_respondents_open_source_question) * 100:.2f}%")
```

```python
country_counts = data['Country'].value_counts().head(20)

plt.figure(figsize=(12, 8))
sns.barplot(x=country_counts.values, y=country_counts.index, palette='crest')
plt.title('Top 20 Countries by Number of Respondents')
plt.xlabel('Number of Respondents')
plt.ylabel('Country')
plt.show()

print("Top 20 Countries by Respondents:\n", country_counts)
```

```python
# Column: FormalEducation
education_counts = data['FormalEducation'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=education_counts.values, y=education_counts.index, palette='mako')
plt.title('Distribution of Formal Education Levels')
plt.xlabel('Number of Respondents')
plt.ylabel('Formal Education')
plt.show()

print("Formal Education Distribution:\n", education_counts)
```

```python
# Column: LanguageWorkedWith (This is a multi-select column, requiring splitting)
# It contains comma-separated values, e.g., 'Python;JavaScript;HTML'.
if 'LanguageWorkedWith' in data.columns:
    # Drop rows where 'LanguageWorkedWith' is NaN to avoid errors
    languages_raw = data['LanguageWorkedWith'].dropna()
    all_languages = []
    for languages_str in languages_raw:
        all_languages.extend([lang.strip() for lang in languages_str.split(';')])

    language_counts = pd.Series(all_languages).value_counts().head(10)

    plt.figure(figsize=(12, 7))
    sns.barplot(x=language_counts.values, y=language_counts.index, palette='rocket')
    plt.title('Top 10 Most Popular Programming Languages (Worked With)')
    plt.xlabel('Number of Developers')
    plt.ylabel('Language')
    plt.show()

    print("Top 10 Programming Languages:\n", language_counts)
else:
    print("Column 'LanguageWorkedWith' not found. Cannot perform this analysis.")

```

```python
# Analysis 4.3: Relationship between Years Coding and Converted Salary (using a scatter plot)
print("\n--- Analysis 4.3: Years Coding vs. Converted Salary ---")
# Columns: YearsCoding, ConvertedSalary
# Convert 'YearsCoding' to numeric, handling various string formats.
# And drop NaNs for 'ConvertedSalary'

df_salary_years = data[['YearsCoding', 'ConvertedSalary']].dropna()

# Function to convert 'YearsCoding' strings to a numeric value
def convert_years_coding_to_numeric(years_str):
    if pd.isna(years_str):
        return np.nan
    years_str_lower = str(years_str).lower()

    if 'less than 1 year' in years_str_lower:
        return 0.5
    elif ' or more years' in years_str_lower:
        # Extract the number before " or more years"
        try:
            return float(years_str_lower.split(' ')[0])
        except ValueError:
            return np.nan
    elif '-' in years_str_lower:
        # Handle ranges like '6-8 years' by taking the midpoint
        try:
            parts = years_str_lower.replace(' years', '').split('-')
            return (float(parts[0]) + float(parts[1])) / 2
        except ValueError:
            return np.nan
    else:
        # Handle single year values like '1 year', '2 years', etc.
        try:
            return float(years_str_lower.replace(' years', '').replace(' year', '').strip())
        except ValueError:
            return np.nan # Fallback for any unhandled string format

# Apply the conversion function to create the new numeric column
df_salary_years['YearsCodingNumeric'] = df_salary_years['YearsCoding'].apply(convert_years_coding_to_numeric)

# Drop any rows where 'YearsCodingNumeric' could not be converted (i.e., is NaN)
df_salary_years = df_salary_years.dropna(subset=['YearsCodingNumeric'])

# Cap the salary for better visualization if there are extreme outliers
# For example, cap at the 99th percentile to focus on the majority of data
salary_cap = df_salary_years['ConvertedSalary'].quantile(0.99)
df_salary_years_capped = df_salary_years[df_salary_years['ConvertedSalary'] <= salary_cap]

plt.figure(figsize=(12, 7))
sns.scatterplot(x='YearsCodingNumeric', y='ConvertedSalary', data=df_salary_years_capped, alpha=0.3, color='purple')
plt.title('Years Coding vs. Converted Annual Salary (Capped at 99th Percentile)')
plt.xlabel('Years Coding (Numeric)')
plt.ylabel('Converted Annual Salary (USD)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
