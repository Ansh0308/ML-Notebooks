```python
#funtion normalization
def feature_nomalization(x):
  mean=np.mean(x,axis=0)
  std=np.std(x,axis=0)
  x_normal=(x-mean)/std
  return x_normal,mean,std
```

```python
data_nor=data.values
m=data_nor.shape[0]
#take feature vectors
x2=data_nor[:,0:2].reshape(m,2)
x2_norm,mean,std=feature_nomalization(x2)

y2=data_nor[:,2:3].reshape(m,1)#price col
```

```python
theta_array=np.zeros((3,1))
```

```python
#linear regression
def hypothesis(theta_array,x1,x2):
  return theta_array[0]+theta_array[1]*x1 +theta_array[2]*x2

def cost_funtion(theta_array,x1,x2,y,m):
  sum=0
  for i in range(m):
    sum=sum+((theta_array[0]+theta_array[1]*x1[i]+theta_array[2]*x2[i])-y[i])**2

  return sum/(2*m)

def gradiant_descent(theta_array,x1,x2,y,m,alpha):
  summation_0=0
  summation_1=0
  summation_2=0
  for i in range(m):
    summation_0=summation_0+((theta_array[0]+theta_array[1]*x1[i]+theta_array[2]*x2[i])-y[i])
    summation_1=summation_1+x1[i]*((theta_array[0]+theta_array[1]*x1[i]+theta_array[2]*x2[i])-y[i])
    summation_2=summation_2+x2[i]*((theta_array[0]+theta_array[1]*x1[i]+theta_array[2]*x2[i])-y[i])
  new_theta0=theta_array[0]-(alpha/m)*summation_0
  new_theta1=theta_array[1]-(alpha/m)*summation_1
  new_theta2=theta_array[2]-(alpha/m)*summation_2
  updated_new_theta=[new_theta0,new_theta1,new_theta2]
  return updated_new_theta

def training(x1,x2,y,alpha,iters,theta_array,m):

  cost_values=[]

  for i in range(iters):
    theta_array=gradiant_descent(theta_array,x1,x2,y,m,alpha)
    cost_values.append(cost_funtion(theta_array,x1,x2,y,m))

  return cost_values,theta_array

```

```python
alpha=0.01
itres=500
cost_val,theta_array=training(x2_norm[:,0:1],x2_norm[:,1:2],y2,alpha,itres,theta_array,m)
x_axis=np.arange(0,len(cost_val),step=1)
plt.plot(x_axis,cost_val)
plt.xlabel("ites")
plt.ylabel("Cost_val")
plt.title("Loss graph")
plt.show()
print(theta_array)
```

```python
import numpy as np

def hypothesis(theta_array, X):
    """
    Compute the hypothesis for any number of features.
    theta_array: Array of shape (n+1,) [intercept + n features]
    X: Feature matrix of shape (m, n), excluding intercept
    Returns: predictions of shape (m,)
    """
    return theta_array[0] + np.dot(X, theta_array[1:])

def cost_function(theta_array, X, y, m):
    """
    Compute the cost for linear regression.
    theta_array: parameter vector (n+1,)
    X: Feature matrix (m, n)
    y: Target values (m,)
    m: Number of training examples
    Returns: Scalar cost
    """
    predictions = hypothesis(theta_array, X)
    return np.sum((predictions - y) ** 2) / (2 * m)

def gradient_descent(theta_array, X, y, m, alpha):
    """
    Perform one step of gradient descent for any number of features.
    theta_array: parameter vector (n+1,)
    X: Feature matrix (m, n)
    y: Target values (m,)
    m: Number of training examples
    alpha: Learning rate
    Returns: Updated parameter vector (n+1,)
    """
    predictions = hypothesis(theta_array, X)
    errors = predictions - y
    new_theta_array = theta_array.copy()
    new_theta_array[0] -= (alpha / m) * np.sum(errors)
    for j in range(1, len(theta_array)):
        new_theta_array[j] -= (alpha / m) * np.dot(errors, X[:, j - 1])
    return new_theta_array

def training(X, y, alpha, iters, theta_array, m):
    """
    Train linear regression parameters with any number of features.
    X: Feature matrix (m, n)
    y: Target values (m,)
    alpha: Learning rate
    iters: Number of iterations
    theta_array: Initial parameter vector (n+1,)
    m: Number of training examples
    Returns: List of cost values, final parameter vector
    """
    cost_values = []
    for i in range(iters):
        theta_array = gradient_descent(theta_array, X, y, m, alpha)
        cost_values.append(cost_function(theta_array, X, y, m))
    return cost_values, theta_array
```
